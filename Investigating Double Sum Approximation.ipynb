```python
import torch
import matplotlib.pyplot as plt
import numpy
```

# What do these images actually look like?

```python
images = torch.load('/Users/sragavan/ele538bfinal/images.pt')
labels = torch.load('/Users/sragavan/ele538bfinal/labels.pt')
```

```python
images.shape, labels.shape
```

```python
for i in range(128):
    img = images[i].squeeze().numpy()
    lbl = labels[i].squeeze().numpy()
    print(i, lbl)
    plt.imshow(img, cmap='Greys')
    plt.show()
    print('------------')
```

```python
for i in range(128, 256):
    img = images[i].squeeze().numpy()
    lbl = labels[i].squeeze().numpy()
    print(i, lbl)
    plt.imshow(img, cmap='Greys')
    plt.show()
    print('------------')
```
